```python
import numpy as np
import pandas as pd

a=pd.read_csv("BayesData.csv")

ab = np.array(a)

a1=[]; b1=[]; c1=[]; d1=[]; a11=[]; b11=[]; c11=[]; d11=[]
b = ab[:-1,:]


print("Training Data Set to :\n", b)

ay=sum(b[:,-1]=="Y")
an=sum(b[:,-1]=="N")

O = ab[-1]
print("Test Data Set to = ", O)

for i in range(len(b)):
  if (b[i,-1]=="Y"):
    a1.append(b[i,0]==O[0]);f1=np.sum(a1)/ay
    b1.append(b[i,1]==O[1]);f2=np.sum(b1)/ay
    c1.append(b[i,2]==O[2]);f3=np.sum(c1)/ay
    d1.append(b[i,3]==O[3]);f4=np.sum(d1)/ay
  
  else:
    a11.append(b[i,0]==O[0]);f11=(np.sum(a11))/an
    b11.append(b[i,1]==O[1]);f21=np.sum(b11)/an
    c11.append(b[i,2]==O[2]);f31=np.sum(c11)/an
    d11.append(b[i,3]==O[3]);f41=np.sum(d11)/an

py=ay/len(b)
pn=an/len(b)

pyes=f1*f2*f3*f4*py
pno=f11*f21*f31*f41*pn

poyes=pyes*100/(pyes+pno)
pono=pno*100/(pyes+pno)

print("p(Yes) = ",poyes)
print("P(No) = ",pono)

if poyes > pono:
  print("Probabiltiy of event occurence = ", poyes)
else:
  print("Probability of non-occurence of event = ", pono)
```
