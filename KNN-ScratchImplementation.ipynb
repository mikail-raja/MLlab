```python
import pandas as pd
import numpy as np
import math
import collections
```

```python
data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/ML_Lab/IRIS.csv")
print(data)
```

```python
x = np.array(data)[:, :-1]
y = np.array(data)[:, -1]

for idx, val in enumerate(y):
  if val == "Iris-virginica":
    y[idx] = 2
  if val == "Iris-versicolor":
    y[idx] = 1
  if val == "Iris-setosa":
    y[idx] = 0

# print(x)
# print(y)
```

```python
def euclideandist(inp):
  distance = []
  for i in range(x.shape[0]):
    dist = (x[i][0] - inp[0]) **2 + (x[i][1] - inp[1]) **2 + (x[i][2] - inp[2]) **2 + (x[i][3] - inp[3]) **2
    distance.append(math.sqrt(dist))
  return distance
```

```python
def knn(inp):
  dist = euclideandist(inp)
  distances = zip(dist, y)
  print(distances)
```

```python
examplex = np.array([5, 4, 2, 0.2])
knn(examplex)
```
