# Write a program to implement naive bayesian classifier using sample data from csv file. compute accuracy of the classifier.

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("BayesData.csv")
print(df)
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
df['WC']= label_encoder.fit_transform(df['WC'])
df['HU']= label_encoder.fit_transform(df['HU'])
df['WT']= label_encoder.fit_transform(df['WT'])
df['WC'].unique()
df['HU'].unique()
df['WT'].unique()
print(df)
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb=GaussianNB()
gnb.fit(x_train,y_train)
```

```python
y_pred=gnb.predict(x_test)
y_pred
```

```python
from sklearn import  metrics
print("Naive Bayesian accuracy is :",metrics.accuracy_score(y_test,y_pred)*100)
```
